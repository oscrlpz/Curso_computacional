### [1] Crear gas ideal

Genero el tipo `Particula`:

```julia
type Particula
    x::Vector{Float64}
    v::Vector{Float64}
    m::Float64
end
```

Posteriormente defino la función `mover!`:

```julia
function mover!(p::Particula, dt)  
    p.x += p.v*dt ## Como p.x es un vector, el algebra de dicho vector me permite hacer esto y 
                  #no tener que irme entrada por entrada
end
```

```julia
p1=Particula([0,1],[1,2],10)
```

```julia
mover!(p1,1)
```

El tipo `Gas` se compone de un entero $N$ que define cuantas partículas contiene dicho gas y un vector de $N$ entradas, donde cada entrada es un tipo `Particula`.

```julia
type Gas
    N::Int
    particulas::Vector{Particula}
end
```

Como el tipo `Gas` define un constructor, se aprovechará esto para definir un segundo método para dicho constructor. Éste nuevo método lo que permitirá será que únicamente se tendrá que introducir el vector de `g.particulas` en el argumento del gas y automáticamente se generará el valor de la entrada del entero `g.N`.

```julia
function Gas(particulas::Vector{Particula})
    Gas(length(particulas), particulas)
end
```

La siguiente función hace uso de la función `rand()` para generar una `Particula` con entradas aleatorias. A dicha función la llame `rndmp()`.

```julia
function rndmp()
    Particula([rand(),rand()],[rand()*(-1)^(rand(0:1)),rand()*(-1)^rand(0:1)],rand())
end
```

Es necesario que las velocidades puedan tener valores negativos, es por eso que se mete el signo aleatorio $(-1)^{rand(0:1)}$.

```julia
rndmp()
```

La función `agregar` se muestra a continuación y lo que hace es agregar una `Particula` $p$ a un `Gas` $g$. 

```julia
function agregar(g::Gas,p::Particula)
    q=g.particulas[1:g.N]
    push!(q,p)
    g.particulas=q
    g.N+=1
end
```

Para generar un gas lo que se tendrá que hacer es ir agregando $N$ `Particulas` aleatorias a un `Gas`. Para hacer esto más sencillo, se definirá un segundo método para la función `agregar`, donde cuando se llame a dicha función y únicamente se introduzca como argumento un `Gas` $g$, entonces, agregue a dicho `Gas` $g$ una `Particula` aleatoria.

```julia
function agregar(g::Gas)
    q=g.particulas[1:g.N]
    push!(q,rndmp())
    g.particulas=q
    g.N+=1
end
```

La función `generar_gas(N)` hace uso de la función `agregar` para generar un `Gas` con $N$ `Particulas` aleatorias.

```julia
function generar_gas(N)
    g=Gas(0,[])
    for n=1:N
        agregar(g)
    end
    g
end
```

Para lograr que se muevan las particulas del `Gas` se define un segundo método para la función `mover!`, que permita mover cada una de las particulas de dicho gas, aplicando la función `mover!` a cada una de ellas.

```julia
function mover!(g::Gas,dt)
    for n=1:g.N
        mover!(g.particulas[n],dt)
    end
end
```

Los valores de dicho gas unicamente van de $0$ a $1$, si se quisiera que fueran diferentes solo habría que generar una función `rndmp()` que tuviera otros límites. Por el momento esto no tiene ningún problema. 

A continuación se muestra como sirve todo lo mostrado anteriormente. Para empezar se genera un `Gas` llamado `gas1`.

```julia
gas1=generar_gas(100)
```

Y se le aplica la función `mover!` para un tiempo $dt$ arbitrario.

```julia
mover!(gas1,0.125)
```

Véase como las posiciones de las particulas de dicho `Gas` se vieron modificadas.

```julia
gas1
```

###  [2] Condiciones en la frontera

Es facil darse cuenta que lo que se hace al generar un gas es generar $N$ partículas encerradas en un cuadro de $1 \times 1$. A partir de eso se puede considerar como interactuan las partículas con las fronteras.

Lo que haré será meter las condiciones de la frontera dentro de la función que define el movimiento de las particulas. Por ejemplo en el caso de la función `mover` que ya se definió, si bien las particulas inician en un cuadrado $1 \times 1$ cuando pase el tiempo lo único que ocurrira será que estas se expandirán al infinito sin ninguna restricción.


Para simplificar las cosas se considerará que las fronteras siempre serán la caja $ 1 \times 1$ dentro de la cual se generan las párticulas. Si se quiere cambiar las dimensiones de la caja unicamente bastaría hacer que los números aletorios generados para la posición fueran más allá del intervalo $[0,1]$. No considero que esto sea esencial asi que por eso no lo implementaré por el momento y me limitaré a cajas de partículas de $ 1 \times 1$.

A continuación se presenta los diferentes tipos de frontera que pude meter.

--------------------------
**A. Sale por un lado, regresa por el otro**

En este caso cuando se mueva una particula y salga de la caja, está se regresara por el otro lado. 

Es necesario hacer que los pasos de la función `mover` correspondiente sean lo suficientemente pequeños como para que pueda seguir lo más cercanamente posible a la partícula.

La función mover con estas condiciones a la frontera la llame `mover_inf` y la muestro a continuación:

```julia
function mover_inf!(p::Particula,dt)
    mover!(p::Particula,dt)
     for i=1:2
        if p.x[i] > 1 
            p.x[i]=p.x[i]-1
        elseif p.x[i]<0
            p.x[i]=p.x[i]+1
                
        end
    end
end
```

Para ver que realmente esta partícula se comporta como se desea, hay que definir la función `trayectoria_inf` que devuelve la posición de la partícula moviendose de acuerdo a la función `mover_inf` a cada intervalo de tiempo $dt$ bajo un intervalo total de tiempo $T$.

```julia
function trayectoria_inf(p::Particula,dt,T)
    t=dt
    r=Vector{Float64}[]
    while t<T
        mover_inf!(p,dt)
        r=push!(r,p.x)
        t+=dt
    end
    r
end
```

Al final se tiene un arreglo con todas las posiciones de la partícula el cual se puede graficar asi:

```julia
r=trayectoria_inf(rndmp(),.015625,10)

using PyPlot
x=[r[i][1] for i in 1:length(r)];
y=[r[i][2] for i in 1:length(r)];
plot(x, y, linewidth=1.0)
```

Las grandes lineas horizontales o verticales que atraviesan toda la caja aparecen cada vez que la partícula toca la frontera y brinca al otro lado de la caja.

Para aplicar este mismo movimiento a cada una de las partículas de un gas, se hace algo análogo a lo que ya se hizo más arriba con la función `mover`. Así, se escribe la función `mover_inf!` aplicada a un `Gas` como:

```julia
function mover_inf!(g::Gas,dt)
    for n=1:g.N
        mover_inf!(g.particulas[n],dt)
    end
end
```

Más adelante visualizaré como evolucióna el gas con condiciones a la frontera.



-----------------
**B. Colisión inelástica**

En este caso cuando se llega a las fronteras, la partícula pierde su energía cinética y se detiene, es decir las componentes de su vector velocidad se vuelven cero. De modo que su correspondiente función mover es:

```julia
function mover_inelas!(p::Particula,dt)
    mover!(p::Particula,dt)
     for i=1:2
        if p.x[i] > 1
            p.x[i]=1
            p.v[1:2]=0
        elseif p.x[i]<0
            p.x[i]=0
            p.v[1:2]=0
        end
    end
end
```

```julia
function trayectoria_inelas(p::Particula,dt,T)
    t=dt
    r=Vector{Float64}[]
    while t<T
        mover_inelas!(p,dt)
        r=push!(r,p.x)
        t+=dt
    end
    r
end
```

```julia
r=trayectoria_inelas(rndmp(),.5/128, 10)
using PyPlot
x=[r[i][1] for i in 1:length(r)];
y=[r[i][2] for i in 1:length(r)];
plot(x, y, linewidth=1.0)
```

Claramente este es el comportwamiento deseado ya que la partícula se mueve en movimiento rectilíneo uniforme hasta topar con pared.

Ahora, para mover al gas se escribe:

```julia
function mover_inelas!(g::Gas,dt)
    for n=1:g.N
        mover_inelas!(g.particulas[n],dt)
    end
end
```

-------------------
**C. Colisión elástica**

En este caso cuando se llega a las fronteras el vector velocidad y el vector posición cambian de dirección de manera especular.

```julia
function mover_elas!(p::Particula,dt)
    mover!(p::Particula,dt)
     for i=1:2
        if p.x[i] > 1 || p.x[i]<0
           p.v[i]=-p.v[i]
        end
    end
end
```

```julia
function trayectoria_elas(p::Particula,dt,T)
    t=dt
    r=Vector{Float64}[]
    while t<T
        mover_elas!(p,dt)
        r=push!(r,p.x)
        t+=dt
    end
    r
end
```

```julia
r=trayectoria_elas(rndmp(),.5/128, 100)
using PyPlot
x=[r[i][1] for i in 1:length(r)];
y=[r[i][2] for i in 1:length(r)];
plot(x, y, linewidth=1.0)
```

En ésta gráfica se ve claramente como la partícula rebota sobre las paredes que van de $0$ a $1$ en $x$ y $y$.

Ahora aplicando este movimiento con fronteras elásticas a las partículas del gas se tiene que:

```julia
function mover_elas!(g::Gas,dt)
    for n=1:g.N
        mover_elas!(g.particulas[n],dt)
    end
end
```

### [3] Visualizar gas

Si se quiere graficar las posiciones de las particulas del gas es necesario extraer únicamente los valores de $x$ y $y$ de dicho gas. Esto se logra con la siguiente función.

```julia
function posiciones(g::Gas)
    y=Any[]
    for n=1:g.N
        x=g.particulas[n].x 
        push!(y,x)
    end
    y
end
```

**A. Gas elástico**

Por ejemplo, si genero un gas1:

```julia
gas1=generar_gas(100)
```

Uso la función `posiciones` para obtener los vectores posición de las partículas del gas1 y gráfico el arreglo extraido.

```julia
r=posiciones(gas1)
using PyPlot
x=[r[i][1] for i in 1:length(r)];
y=[r[i][2] for i in 1:length(r)];
scatter(x, y)
```

Si quisiera graficar las trayectorias de cada una de las partículas sería un desorden de lineas que no me dirií absolutamente nada, seria más útil si pudiera hacer una animación que muestre la evolución temporal del gas. Como no me dio tiempo de aprender a hacer eso lo que haré será definir una función evolución que me da el estado del final del sistema tras haber transcurrido un tiempo $T$, donde la trayectoria del sistema del gas se calcula a intervalos de tiempo $dt$.

La función `evolucion_gas_elas` se muestra a continuación:

```julia
function evolucion_gas_elas(g::Gas,dt,T)
    t=dt
    while t<T
        mover_elas!(g::Gas,dt)
        t=t+dt
    end
    
end
```

Aquí elegí la evolución temporal de un gas con colisiones elásticas.

Ahora observo como cambiaron de posición las particulas del gas1 anterior después de un tiempo $T=100$ a intervalos $dt=0.5/128$.

```julia
evolucion_gas_elas(gas1,0.5/128,100)
```

```julia
r=posiciones(gas1)
using PyPlot
x=[r[i][1] for i in 1:length(r)];
y=[r[i][2] for i in 1:length(r)];
scatter(x, y)
```

Nótese como las particulas se mueven pero jamás salen de la caja $1 \times 1$.b
