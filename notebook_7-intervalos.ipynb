¿Cuál es la diferencia entre el cálculo mediante intervalos y el cálculo original que hicimos?

Haciendo la suma $$\sum^\infty _{n=1} \frac{1}{n^2} = \frac{\pi^2}{6},$$ pero donde $n$ será tomado como intervalo.

Fue necesario introducir una fnción adicional al módulo, la cual añadiera un nuevo método al constructor `Intervalo`. Dicho constructor tiene la finalidad de hacer que cuando el argumento de `Intervalo` sea un solo número te lo vuelva un intervalo donde su cota inferior es el redondeo para abajo de dicho número y la cota superior sea el redondeo para arriba del mismo número.

Esto es, supongase que se escribe `Intervalo(x)`, lo que la maquina devolvera será un tipo `Intervalo` $(x_a, x_b)$, donde $x_a$ sea el `BigFloat` redondeado para abajo de $x$ y $x_b$ sea el `BigFloat` redondeado para arriba de $x$.

El codigo del módulo presentado anteriorente ya incluía dicha función.

Realizando con intervalos la suma ya mencionada:

```julia
s=sum([Intervalo(1)/(Intervalo(i)*Intervalo(i)) for i=1:100])
```

Hay que recordar que hay que realizar la corrección al añadir las colas, lo cual se hace a continuación:

```julia
r=Intervalo(s.a+1/101, s.b+1/100)
```

Para comprobar que efectivamente el valor de $\frac{\pi^2}{6}$ se encuentra en el intervalo dado por $r$:

```julia
pi^2/6>r.a && pi^2/6<r.b
```

Lo cual resulta verdadero.

(iii) ¿Qué representa el resultado?

Son los valores máximos y mínimos que tiene la función $p_1(x)$ en el intervalo elegido. Esto es, parecería ser que dado un dominio para la función $p_1(x)$ se devuelve el intervalo del codominio.


Por ejemplo, para el intervalo $[3,5]$, utilizando dicho intervalo para la función $p_1$ se llega a que:

```julia
p=p1(Intervalo(3,5))
```

Por otro lado si hacemos un arreglo con los valores de $p_1$ usando valores de $x$ dentro del intervalo $[3,5]$ se obtiene:

```julia
R=[p1(n) for n=3:0.0625:5]
```

Además se puede obtener cual fue el máximo y mínimo que se obtuvo en dicho intervalo. 

```julia
println("El minimo es: ", minimum(R), " y el máximo es: ", maximum(R))
```

Y de esta manera se sabe que dicho minimo y maximo corresponden a las cotas del intervalo del codominio.

Para ver esto más claramente se peude graficar con las siguientes funciones.

```julia
using PyPlot
```

La función `p1_graf_intervalo` gráfica la curva de la función $p_1$ y en un cuadrado amarillo indica el área delimitada por el intervalo que se tomó para $x=[a,b]$ y el intervalo resultado de $p_1(x)$, donde $x$ es tipo `Intervalo`.

```julia
function p1_graf_intervalo(a,b)
    x=[n for n=-3:0.0625:6]
    f=[p1(n) for n=-3:0.0625:6]
    p=p1(Intervalo(a,b))
    plot(x,f)
    PyPlot.fill_between([a,b],float64(p.a),float64(p.b),color="yellow")
end
```

Una segunda grafica se construye  de manera similar, sólo que ahora se usa un cuadrado verde cuya área s eencuentra delimitada por el intervalo del dominio y por el intevarvalo real del codominio.

```julia
function p1_graf_dominio(a,b)
    x=[n for n=-3:0.0625:6]
    f=[p1(n) for n=-3:0.0625:6]
    R=[p1(n) for n=a:0.0625:b]
    
    plot(x,f)
   
    PyPlot.fill_between([a,b],minimum(R),maximum(R),color="green")
end
```

En las siguientes dos gráficas se observa que para $[3,5]$ el resultado, usando tipos `Intervalo` coincide con el codominio de la función.

```julia
p1_graf_intervalo(3,5)
```

```julia
p1_graf_dominio(3,5)
```

```julia
function q1(x::Real)
    x^2-3*x+2
end

function q1(x::Intervalo)
    x^2-Intervalo(3)*x+Intervalo(2)
end

function q1_graf_intervalo(a,b)
    x=[n for n=-3:0.0625:6]
    f=[q1(n) for n=-3:0.0625:6]
    q=q1(Intervalo(a,b))
    plot(x,f)
    PyPlot.fill_between([a,b],float64(q.a),float64(q.b),color="yellow")
end

function q1_graf_dominio(a,b)
    x=[n for n=-3:0.0625:6]
    f=[q1(n) for n=-3:0.0625:6]
    R=[q1(n) for n=a:0.0625:b]
    
    plot(x,f)
   
    PyPlot.fill_between([a,b],minimum(R),maximum(R),color="green")
end
```

(iii) Pruébalo con otros polinomios.

Utilizaré el polinomio que se ocupó en el *notebook 4*, este es: $$p_3(x)=(1-x)^6 ,$$ donde su forma desarrollada es: $$    q_3(x)= x^6-6(x^5)+15(x^4)-20(x^3)+15(x^2)-6x+1$$


Escogí éste polinomio ya que, recordando el notebook 4, dicho polinomio se ocupó para demostrar que la forma desarrollada $q_3$ generaba mucho ruido en la zona cerca al mínimo. Precisamente ese ruido se corregía si se ocupaba la forma compacta $p_3$.

A continuación se muestra las gráficas obtenidas para este polinomio expresado en ambas formas.

```julia
function p3(x::Real)
    (1-x).^6
end

function p3(x::Intervalo)
    p3r=(Intervalo(1)-x)^6
    min_p3=minimum([p3(n) for n=-1:0.0625:1]) ###Para hallar el mínimo de la función###
    Intervalo(max(p3r.a,min_p3), p3r.b ) #### Aquí se mete la condición a las cotas###
end

function p3_graf_intervalo(a,b,minp,maxp)
    x=[n for n=minp:0.0625/512:maxp]
    f=[p3(n) for n=minp:0.0625/512:maxp]
    p=p3(Intervalo(a,b))
    plot(x,f)
    PyPlot.fill_between([a,b],float64(p.a),float64(p.b),color="yellow")
end
```

Primero gráfico lejos de la zona del valor mínimo. Y lo que se observa a continuación es lo esperado, el intervalo de resultado corresponde al codominio de la función.

```julia
p3_graf_intervalo(8,9,-10,12)
```

La siguiente gráfica muestra que en la región cercana al mínimo. Ya se llevó a cabo la corrección, por lo cual coincide el intervalo resultado con el codominio.

```julia
p3_graf_intervalo(-2,3,-4,5)
```

```julia
function q3(x::Real)
    x.^6-6*(x.^5)+15*(x.^4)-20*(x.^3)+15*(x.^2)-6*x+1
end

function q3(x::Intervalo)
    q3r=x^6-Intervalo(6)*(x^5)+Intervalo(15)*(x^4)-Intervalo(20)*(x^3)+Intervalo(15)*(x^2)-Intervalo(6)*x+Intervalo(1)
    min_q3=minimum([q3(n) for n=-1:0.0625:1]) ###Para hallar el mínimo de la función###
    Intervalo(max(q3r.a,min_q3), q3r.b ) #### Aquí se mete la condición a las cotas###
end

function q3_graf_intervalo(a,b,minp,maxp)
    x=[n for n=minp:0.0625/1028:maxp]
    f=[q3(n) for n=minp:0.0625/1028:maxp]
    p=q3(Intervalo(a,b))
    plot(x,f)
    PyPlot.fill_between([a,b],float64(p.a),float64(p.b),color="yellow")
end
```

Ahora lo que se hará es ver como la forma desarrollada se comporta en la región cercana al mínimo.

La gráfica se muestra a continuación y se observa como ahoravel intervalo resultado es más grande que el codominio. Esto reafirma la idea de que el tipo `Intervalo`sirve para diagnozticar que tanto rudo se mete cuando se calcula una función.
